**SELEÇÃO DO DATASET:**

 A base de dados escolhida foi o ChEMBL (https://www.ebi.ac.uk/chembl/). Trata-se de um banco de dados de pequenas moléculas bioativas semelhantes a drogas, que contém estruturas 2-D, propriedades calculadas (por exemplo, logP, peso molecular, parâmetros de Lipinski, etc.) e bioatividades abstraídas (por exemplo, constantes de ligação, farmacologia e dados ADMET). Os dados são resumidos e curados da literatura científica primária e cobrem uma fração significativa da SAR e da descoberta de medicamentos modernos.

Buscando datasets do alvo: **"Trypanosoma"**

```python
alvo = new_client.target
pesquisa_alvo = alvo.search('Trypanosoma')
ds = pd.DataFrame.from_dict(pesquisa_alvo)
ds
```

Buscando no dataset um alvo em específico:

```python
# Definindo o alvo a ser procurado:
alvo = "Trypanosoma"

# Verificando se algum elemento em pref_name contém este alvo:
contains_alvo = ds['pref_name'].str.contains(alvo)

# Obtendo os índices das linhas com o alvo definido:
indices_com_alvo = ds[contains_alvo].index.tolist()

if contains_alvo.any():
    print(f"Pelo menos um elemento contém o termo: {alvo}")
    print(f"Índices das linhas com o termo '{alvo}': {indices_com_alvo}")
else:
    print(f"Nenhum elemento contém o termo: {alvo}")
```

Convertendo valores de IC50 para uma unidade padrão de concentração (M)

```python
# prompt: select rows from df['target_chembl_id'] if in this list: 'CHEMBL612893','CHEMBL3797017','CHEMBL612644'

ensaios = ds[ds['target_chembl_id'].isin(['CHEMBL368', 'CHEMBL612849', 'CHEMBL612851', 'CHEMBL612348', 'CHEMBL613768'])]
ensaios
```

```python
indices_com_ensaio = ensaios.index
indices_com_ensaio
```

```python
# Criando uma lista para armazenar os DataFrames individuais (Necessário apenas na primeira execução!)

dfs = []




# Iterando sobre os diferentes índices:

for i in indices_com_ensaio :

    df_nM_i = []

    df_uM_i = []

    df_mM_i = []

    df_M_i = []

    ds_selecionado_i = ds.target_chembl_id[i]




    # Filtrando os compostos bioativos com dados de IC50, em unidades de nM, para cada índice:

    atividade = new_client.activity

    resultado_nM = atividade.filter(target_chembl_id=ds_selecionado_i).filter(standard_type="IC50").filter(units="nM")




    # Filtrando os compostos bioativos com dados de IC50, em unidades de uM, para cada índice:

    resultado_uM = atividade.filter(target_chembl_id=ds_selecionado_i).filter(standard_type="IC50").filter(units="uM")




    # Filtrando os compostos bioativos com dados de IC50, em unidades de mM, para cada índice:

    resultado_mM = atividade.filter(target_chembl_id=ds_selecionado_i).filter(standard_type="IC50").filter(units="mM")




    # Filtrando os compostos bioativos com dados de IC50, em unidades de M, para cada índice:

    resultado_M = atividade.filter(target_chembl_id=ds_selecionado_i).filter(standard_type="IC50").filter(units="M")




    # Criando um DataFrame para cada unidade:

    df_nM_i = pd.DataFrame.from_dict(resultado_nM)

    df_uM_i = pd.DataFrame.from_dict(resultado_uM)

    df_mM_i = pd.DataFrame.from_dict(resultado_mM)

    df_M_i = pd.DataFrame.from_dict(resultado_M)




    # Convertendo cada DataFrame para uma unidade padrão(M):

    if not df_nM_i.empty and 'value' in df_nM_i:

        df_nM_i['value'] = df_nM_i['value'].astype(float)

        df_nM_i['value'] *= 1e-9

    else:

        pass




    if not df_uM_i.empty and 'value' in df_uM_i:

        df_uM_i['value'] = df_uM_i['value'].astype(float)

        df_uM_i['value'] *= 1e-6

    else:

        pass




    if not df_mM_i.empty and 'value' in df_mM_i:

        df_mM_i['value'] = df_mM_i['value'].astype(float)

        df_mM_i['value'] *= 1e-3

    else:

        pass




    if not df_M_i.empty and 'value' in df_M_i:

        df_M_i['value'] = df_M_i['value'].astype(float)

    else:

        pass




    # Adicionando os DataFrames à lista:

    dfs.append(df_nM_i)

    dfs.append(df_uM_i)

    dfs.append(df_mM_i)

    dfs.append(df_M_i)




# Concatenando os DataFrames individuais em um único DataFrame

df_assays = pd.concat(dfs, ignore_index=True)

df_assays['units'] = 'M'




# Exibindo o DataFrame final

display(df_assays)
```

```python
df_assays["value"].isnull().sum()
```

```python
# Supondo que seu DataFrame seja df_assays
df_assays.dropna(subset=['value'], inplace=True)
```

```python
df_assays
```

```python
# Calculate the percentage of each category in the 'assay_type' column
assay_type_percentages = df_assays['assay_type'].value_counts(normalize=True) * 100
print(assay_type_percentages)
```

```python
# Atribuindo a classe dos compostos
## A Variável de interesse é sempre "standard_value"

bioactivity_class = []
for i in df_assays.standard_value:
    if float(i) >= 10000:
        bioactivity_class.append("Inactive")
    elif float(i) < 1000:
        bioactivity_class.append("Active")
    else:
        bioactivity_class.append("Intermediate")
```

```python
# Visualizando os compostos bioativos
df_assays.molecule_chembl_id
```

```python
# Iterando os compostos bioativos
mol_cid = []
for i in df_assays.molecule_chembl_id:
    mol_cid.append(i)
```

```python
# Imprimindo a variável mol_cid
mol_cid
```

```python
# Iterando canonical smiles a uma lista
canonical_smiles = []
for i in df_assays.canonical_smiles:
    canonical_smiles.append(i)
```

```python
# Iterando para standard_value a uma lista
standard_value = []
for i in df_assays.standard_value:
    standard_value.append(i)
```

```python
# Combinando as quatro variáveis em um mesmo dataframe
dados_tupla = list(zip(mol_cid, canonical_smiles, bioactivity_class, standard_value))
df3 = pd.DataFrame( dados_tupla,  columns=['molecule_chembl_id', 'canonical_smiles', 'bioactivity_class', 'standard_value'])
```
