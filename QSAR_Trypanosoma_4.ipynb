#**Combining the compound fingerprint descriptors dataset and the biological activity dataset (pIC50)**

```python
# Viewing the imported dataset

import pandas as pd
df1 = pd.read_csv("pubchem_descritores_3classes.csv")
display (df1.head())
```

### Treat the descriptors dataset
